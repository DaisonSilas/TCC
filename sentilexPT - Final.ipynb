Importando as Bibliotecas

```python
import nltk
import numpy as np
import pandas as pd
```

Importando o Léxico

```python
sentilexpt = open('SentiLex-lem-PT-melhorado.txt','r', encoding = 'utf-8')
```

Iterar sobre o Léxico

```python
dic_palavra_polaridade = {}
for i in sentilexpt.readlines():
    pos_ponto = i.find('.')
    palavra = (i[:pos_ponto])
    pol_pos = i.find('POL')
    polaridade = (i[pol_pos+4:pol_pos+6]).replace(';',' ')
    dic_palavra_polaridade[palavra] = polaridade
```

Atribuindo valores a cada palavra de acordo com o Sentilex na frase de entrada e retornar o valor final da frase completa 

```python
def Score_sentimento(frase):
    frase = frase.lower()
    l_sentimento = []
    for p in frase.split():
        l_sentimento.append(int(dic_palavra_polaridade.get(p, 0)))
    score = sum(l_sentimento)
    return '{}'.format(score)
    
```

Lendo o arquivo do banco de dados

```python
dados = pd.read_csv("Banco de dados melhorados1.csv", delimiter=";")
```

Usando a função Score_sentimento para colocar no Dataframe uma coluna "Sentimento" com o valor da frese inteira

```python
sentimento = [Score_sentimento(dados["Frase"][ind]) for ind in dados.index]
```

```python
dados["Sentimento"] = sentimento
```

```python
banco = dados
```

Classificando as frases de acordo com os seus valores para definir suas polaridades

```python
def polaridade_numero(pol):
    if int(pol) <= -1:
        return -1
    elif int(pol) == 0:
        return 0
    else: 
        return 1
```

Atribuindo valores numericos ao rotulos das frase para comparar com os valores na def "acuracia"

```python
def polaridade_texto(pol):
    if pol == 'Positivo':
        return 1
    elif pol == 'Neutro':
        return 0
    else:
        return -1
```

Comparação entre os rotulos e os valores das frases

```python
def acuracia(anls, pred):
    if (anls - pred) == 0:
        return 'certo'
    else:
        return 'errado'
```

Organizando no Dataframe as colunas com os seus respectivos valores

```python
polaridade = [polaridade_texto(banco['Nvpolaridade'][ind]) for ind in banco.index]
```

Nvpolaridade é a coluna que contem um rotulo atribuido por nos, com a finalidade de aproximar ainda mais com a polaridade da frase

```python
banco['Nvpolaridade'] = polaridade
```

```python
polaridade_num = [polaridade_numero(banco['Sentimento'][ind]) for ind in banco.index]
```

```python
banco['Sentimento'] = polaridade_num
```

```python
acuracia = [acuracia(banco['Nvpolaridade'][ind], banco['Sentimento'][ind]) for ind in banco.index]
```

```python
acuracia = pd.DataFrame(acuracia)
```

```python
banco['Acuracia'] = acuracia
```

Função para retornar os valores das frases individualmente para analise

```python
def Score_sentimentotest(frase):
    frase = frase.lower()
    l_sentimento = dict()
    for p in frase.split():
        l_sentimento[p] = int(dic_palavra_polaridade.get(p, 0))
    for i in l_sentimento.items():
        return l_sentimento
```
